## 6. Cria uma variável df contendo as colunas <i>country</i>, <i>province</i>, <i>region_1</i> e <i>region_2</i> com registros dos índices 0, 1, 10, 100.

## 8. Crie um DataFrame <i>italian_wines</i> contendo reviews de vinhos feitos na <i>Italy</i>.

## 9. Crie um DataFrame <i>top_oceania_wines</i> contendo todos os reviews de pelo menos 95 (de 100) para vinhos da Australia ou New Zealand

```python
top_oceania_wines = reviews.loc[(reviews.points >= 95) & reviews.country.isin(['Australia', 'New Zealand'])]
```
